# Definir Funciones Principales
En esta sección, se definen las funciones clave utilizadas en el script `main.py`.

```python
# Definir las funciones principales

def initialize_retriever():
    """Inicializa el objeto Retriever."""
    return Retriever()

def launch_rag_interface(retriever):
    """Lanza la interfaz de chat RAG."""
    rag_chat_interface = RAGChatInterface(retriever)
    rag_chat_interface.launch_interface()
```

# Ejecutar el Flujo Principal
En esta sección, se ejecuta el flujo principal del script, utilizando las funciones definidas anteriormente.

```python
import contextlib
import io

# Ejecutar el flujo principal
try:
    # Redirigir la salida estándar y de error a un flujo nulo
    with contextlib.redirect_stdout(io.StringIO()), contextlib.redirect_stderr(io.StringIO()):
        logger.info("[Notebook] Iniciando flujo principal.")

        # Inicializar el objeto Retriever
        retriever = initialize_retriever()

        # Lanzar la interfaz de chat RAG
        launch_rag_interface(retriever)

        logger.info("[Notebook] Flujo principal ejecutado con éxito.")
except Exception as e:
    logger.error(f"[Notebook] Ocurrió un error: {e}")
```
